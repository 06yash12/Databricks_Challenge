```python
# Create widgets
dbutils.widgets.text("source_table", "default.events_table")
dbutils.widgets.dropdown("layer", "bronze", ["bronze", "silver", "gold"])

# Read widget values
source_table = dbutils.widgets.get("source_table")
layer = dbutils.widgets.get("layer")

print("Source Table:", source_table)
print("Layer:", layer)

# What this does :-
# Creates input parameters
# Makes notebook reusable
# Allows job to control bchavior
```

```python
from pyspark.sql import functions as F

if layer == "bronze":
    bronze_df = (
        spark.read.table(source_table)
        .withColumn("ingestion_ts", F.current_timestamp())
    )
    bronze_df.write \
        .format("delta") \
        .mode("overwrite") \
        .saveAsTable("default.bronze_events")

    print("Bronze layer completed successfully")
```
